```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

```python
df = pd.read_csv('energies.txt',names=['F','E','K'],delimiter=',')
k_copy = df['K']*-1
```

```python
plt.figure(figsize=(3, 3), dpi=180)
plt.ylim(-8,8)
plt.plot(df['E'],df['F'],'-b',label='B=0.5(Å). W=1.4(Å). V=13.6(eV)')
plt.axhline(-1,color='Black')
plt.axhline(1,color='Black')
plt.axvline(22,color='red')
plt.xlabel('Energy (eV)')
plt.ylabel('Function')
plt.title("Kronig-Penney Graphical Solution")
plt.legend(fontsize=6)
```
